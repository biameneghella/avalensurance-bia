1-imports + organização

```python

from pathlib import Path
import pandas as pd
import numpy as np


ROOT = Path.cwd().parents[1]  
DATA_DIR = ROOT / "data"

INPUT_FILE  = DATA_DIR / "clinical.csv"
OUTPUT_FILE = DATA_DIR / "clinical_cleaned.csv"

print("Raiz:", ROOT)
print("Entrada:", INPUT_FILE)
print("Saída:", OUTPUT_FILE)
```

2- lendo

```python

df = pd.read_csv(INPUT_FILE)

print("Dimensões originais:", df.shape)
display(df.head())
```

3-limpa nulos

```python

colunas_tratar = ["systolic_bp", "diastolic_bp", "ldl", "hba1c"]

# Contagem antes da limpeza
print("Nulos antes do tratamento:")
display(df[colunas_tratar].isnull().sum())

# Substitui nulos e strings vazias por "indisponivel"
for col in colunas_tratar:
    df[col] = df[col].replace("", np.nan)
    df[col] = df[col].fillna("indisponivel")

# Contagem após a limpeza
print("\nNulos após o tratamento:")
display(df[colunas_tratar].isnull().sum())
```

4- ordenação e remoção por duplicadas

```python


# Ordenar por person_id em ordem crescente
df = df.sort_values(by="person_id", ascending=True)

# Contar quantos IDs duplicados existem antes de remover
duplicados = df.duplicated(subset="person_id").sum()
print(f"Foram encontrados {duplicados} person_id(s) duplicado(s).")

# Remover duplicados mantendo o primeiro
df = df.drop_duplicates(subset="person_id", keep="first").reset_index(drop=True)

print("Dimensões após remoção de duplicados:", df.shape)
```

5- exportação

```python

df.to_csv(OUTPUT_FILE, index=False)
print(f"Arquivo salvo em: {OUTPUT_FILE}")


display(df.head())
```
